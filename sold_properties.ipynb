```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver
driver = webdriver.Chrome(executable_path='/path/to/chromedriver')

# Step 1: Go to the URL
url = "https://www.zillow.com/el-paso-tx-79936/sold/"
driver.get(url)

# Wait for the page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

# Step 2: Click Beds & Baths button
beds_baths_button = driver.find_element(By.XPATH, '//button[@data-test="beds-filters-button"]')
beds_baths_button.click()

# Step 2.1: Select Beds (example: 3+)
beds_button = driver.find_element(By.XPATH, '//button[@value="3"]')
beds_button.click()

# Step 2.3: Select Baths (example: 2+)
baths_button = driver.find_element(By.XPATH, '//button[@value="2"]')
baths_button.click()

# Step 3: Apply button
apply_button = driver.find_element(By.XPATH, '//button[@data-test="close-filters-button"]')
apply_button.click()

# Step 4: Click More button
more_button = driver.find_element(By.XPATH, '//button[@data-test="more-filters-button"]')
more_button.click()

# Step 4.1: Set sqft min (example: 2000)
sqft_min_dropdown = driver.find_element(By.ID, 'sqft-min')
sqft_min_dropdown.send_keys("2000")

# Step 4.2: Set sqft max (example: 2750)
sqft_max_dropdown = driver.find_element(By.ID, 'sqft-max')
sqft_max_dropdown.send_keys("2750")

# Step 5: Apply Button
apply_button = driver.find_element(By.XPATH, '//button[@data-test="close-filters-button"]')
apply_button.click()

# Wait for the filtered results to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "list-card-info")))

# Now you can scrape the data from the page as per your requirements

# Don't forget to close the driver after your scraping is done
driver.quit()
```
