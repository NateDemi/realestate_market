```python
import pandas as pd
import glob

# Path to the directory containing the CSV files
directory_path = 'zillow_data/recently_sold'

# Create a pattern to match all CSV files
pattern = f'{directory_path}/zillow (*.csv'

# List all CSV files in the directory
csv_files = glob.glob(pattern)

# Read and append each CSV file into a DataFrame
df_list = [pd.read_csv(file) for file in csv_files]
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
def extract_info(row):
    address = row['StyledPropertyCardDataArea-c11n-8-84-3__sc-yipmu-0'] if isinstance(row['StyledPropertyCardDataArea-c11n-8-84-3__sc-yipmu-0'], str) else ''
    beds = row['StyledPropertyCardHomeDetailsList-c11n-8-84-3__sc-1xvdaej-0'] if pd.notna(row['StyledPropertyCardHomeDetailsList-c11n-8-84-3__sc-1xvdaej-0']) else ''
    baths = row['StyledPropertyCardHomeDetailsList-c11n-8-84-3__sc-1xvdaej-0 3'] if pd.notna(row['StyledPropertyCardHomeDetailsList-c11n-8-84-3__sc-1xvdaej-0 3']) else ''
    sqft = row['StyledPropertyCardHomeDetailsList-c11n-8-84-3__sc-1xvdaej-0 5'] if pd.notna(row['StyledPropertyCardHomeDetailsList-c11n-8-84-3__sc-1xvdaej-0 5']) else ''
    value = row['PropertyCardWrapper__StyledPriceLine-srp__sc-16e8gqd-1'] if pd.notna(row['PropertyCardWrapper__StyledPriceLine-srp__sc-16e8gqd-1']) else ''
    zipcode = address.split(',')[-1].strip().split(' ')[-1] if address else ''
    sold_date_raw = row['StyledPropertyCardBadge-c11n-8-84-3__sc-6gojrl-0'] if pd.notna(row['StyledPropertyCardBadge-c11n-8-84-3__sc-6gojrl-0']) else ''
    sold_date = sold_date_raw.split(' ')[-1] if 'Sold' in sold_date_raw else ''

    return pd.Series([address, zipcode, beds, baths, sold_date,sqft, value])

# Apply the extraction function
extracted_df = combined_df.apply(extract_info, axis=1)
extracted_df.columns = ['Address', 'Zipcode', 'Beds', 'Baths', 'Sold Date', 'sqft', 'Value']

# Display the extracted information
extracted_df.head()
```

```python
extracted_df.to_csv('./zillow_data/recently_sold.csv')
```
